```python
df_s.columns =['loanamnt', 'terms','Rateofintrst', 'grade', 'Experience',
       'annualinc', 'verificationstatus', 'debtincomeratio', 'numbcredit',
       'totalrevolbal', 'totalcredits', 'totalrecint', 'lastweekpay',
       'totcurrbal', 'MORTGAGE', 'OWN', 'RENT', 'CarsMajorpurchases',
       'Househouseimporovements', 'WeddingVacation', 'creditcard',
       'debtconsolidation', 'medical', 'smallbusiness', 'f']
```

```python
df_s
```

```python
a = df_s['totalrevolbal']
df_s.drop(labels=['totalrevolbal'], axis=1,inplace = True)
df_s.insert(24, 'totalrevolbal', a)
```

```python
df_s.head()
```

```python
list(df_s)
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test  = train_test_split(df_s,test_size = 0.2,random_state=0) # 20% size
```

```python
df_train.head()
```

```python
X_train=df_train.iloc[:,0:24]
y_train=df_train.loc[:,["totalrevolbal"]]

X_test=df_test.iloc[:,0:24]
y_test=df_test.loc[:,["totalrevolbal"]]
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

```python
train_pred = model.predict(X_train)
train_resid = train_pred - y_train["totalrevolbal"]
```

```python
train_rmse = np.sqrt(np.mean(train_resid*train_resid))
train_rmse
```

```python
test_pred2 = model.predict(X_test)
```

```python
test_resid2  = test_pred2 - y_test["totalrevolbal"]
```

```python
test_rmse2 = np.sqrt(np.mean(test_resid2*test_resid2))
test_rmse2
```

```python
test_pred2 = model.predict(X_test)
```

```python
test_resid2  = test_pred2 - y_test["totalrevolbal"]
```

```python
test_rmse2 = np.sqrt(np.mean(test_resid2*test_resid2))
test_rmse2
```
